```python
import os
import cv2
import matplotlib.pyplot as plt

def visualize_bboxes(images_dir, labels_dir, num_samples=3):
    for img_file in os.listdir(images_dir)[:num_samples]:
        img_path = os.path.join(images_dir, img_file)
        label_path = os.path.join(labels_dir, os.path.splitext(img_file)[0] + '.txt')

        img = cv2.imread(img_path)
        if img is None:
            print(f"Не удалось загрузить {img_file}")
            continue

        h, w = img.shape[:2]

        if os.path.exists(label_path):
            with open(label_path) as f:
                for line in f:
                    parts = line.strip().split()
                    if len(parts) >= 5:
                        try:
                            cls = int(parts[0])
                            x_center = float(parts[1])
                            y_center = float(parts[2])
                            width = float(parts[3])
                            height = float(parts[4])

                            x1 = int((x_center - width/2) * w)
                            y1 = int((y_center - height/2) * h)
                            x2 = int((x_center + width/2) * w)
                            y2 = int((y_center + height/2) * h)

                            color = (0, 255, 0) if cls == 0 else (0, 0, 255)
                            label = "dorm_number" if cls == 0 else "spbu_herb"
                            cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
                            cv2.putText(img, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
                        except (ValueError, IndexError) as e:
                            print(f"Ошибка в файле {label_path}, строка: {line.strip()} | Ошибка: {str(e)}")

        plt.figure(figsize=(10, 6))
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(img_file)
        plt.axis('off')
        plt.show()

visualize_bboxes('/content/spbu_passes-4/test/images', '/content/spbu_passes-4/test/labels', num_samples=5)
```

```python
import cv2
import numpy as np
import easyocr

def extract_roi(image_path, bbox):
    x1, y1, x2, y2 = map(int, bbox)
    img = cv2.imread(image_path)
    padding = 5
    roi = img[max(0, y1-padding):min(img.shape[0], y2+padding),
              max(0, x1-padding):min(img.shape[1], x2+padding)]
    return roi
```

```python
def preprocess_digits(roi):
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    kernel = np.ones((2, 2), np.uint8)
    cleaned = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    contrast = clahe.apply(cleaned)

    return contrast
```

```python
reader = easyocr.Reader(['en'], detector=False)

def recognize_easyocr(processed_roi):
    result = reader.readtext(processed_roi, allowlist='0123456789', detail=0)
    return ''.join(result) if result else ""
```

```python
import cv2
import easyocr
from ultralytics import YOLO
import numpy as np

model_yolo = model
reader = easyocr.Reader(['en'], gpu=False)

def preprocess_digits(roi):
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    kernel = np.ones((2, 2), np.uint8)
    return cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

def detect_and_recognize(image_path):
    results = model_yolo(image_path)
    img = cv2.imread(image_path)

    herb_found = False
    number_bbox = None

    for box in results[0].boxes:
        cls_id = int(box.cls)
        cls_name = model_yolo.names[cls_id]
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

        if cls_name == 'spbu_herb':
            herb_found = True
            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(img, 'SPbU Herb', (x1, y1-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)

        elif cls_name == 'dorm_number':
            number_bbox = (x1, y1, x2, y2)
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(img, 'Number', (x1, y1-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

    label = 'SPbU Pass' if herb_found else 'Other Card'
    cv2.putText(img, label, (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 0, 255), 3)

    if herb_found and number_bbox is not None:
        x1, y1, x2, y2 = number_bbox
        pad = 5
        roi = img[max(0,y1-pad):min(img.shape[0],y2+pad),
                  max(0,x1-pad):min(img.shape[1],x2+pad)]
        proc = preprocess_digits(roi)
        ocr_res = reader.readtext(proc, allowlist='0123456789', detail=0)
        number = ''.join(ocr_res) if ocr_res else 'Не распознано'
        cv2.putText(img, f'ID: {number}', (x1, y2+30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        print(f'Обнаружен номер: {number}')
    else:
        print('Либо это не пропуск СПбГУ, либо номер не найден.')

    out = 'result_' + image_path.split('/')[-1]
    cv2.imwrite(out, img)
    print(f'Результат сохранён как {out}')
    return img

```
